```python
import requests
```

```python
url="https://www.zigwheels.com/tata-cars"
r = requests.get(url)
htmlcontent = r.content
htmlcontent
```

```python
from bs4 import BeautifulSoup
soup = BeautifulSoup(htmlcontent,"html.parser")
print(soup.prettify)
```

```python
vehicles=soup.find_all(class_="lnk-hvr fnt-16 b block of-hid h-height ml-0 mb-0-imp")
Car_n=[]
Car_Names=[]
for i in range(0,len(vehicles)):
    Car_n.append(vehicles[i].get_text())
for i in range(0,len(vehicles)):
    Car_Names.append(Car_n[i].strip())
print(Car_Names)
```

```python
len(vehicles)
```

```python
cost=soup.find_all("div",class_="clr-bl")
print(cost)
```

```python
Car_p=[]
Car_Price=[]
for i in range(0,len(vehicles)):
    Car_p.append(cost[i].get_text())
for i in range(0,len(vehicles)):
    Car_Price.append(Car_p[i].strip())
print(Car_Price) 
```

```python
import re
```

```python
Car_Prices = []  # Create an empty list to store results

for i in range(0, len(vehicles)):
    cutter1 = re.findall(r"[\w]+[\.]+[\s]+[0-9]+[\.]+[0-9]+[\s]+[\-]?[\s]+[0-9]+[\.]+[0-9]+[\s]+[\w]+",Car_Price[i])
    Car_Prices.append(cutter1)
Car_Prices    
```

```python
info=soup.find_all(class_="clr-pry fnt-12 pb-10 h-height lh-18 of-hid")
info
```

```python
Car_info=[]
for i in range(0,len(vehicles)):
    Car_info.append(info[i].get_text())
Car_info    
```

```python
Ratings=soup.find_all(class_="r-w fnt-12 rel i-b rt-g")
Rating=soup.find_all(class_="r-w fnt-12 rel i-b rt-lg")
print(Ratings)
Rating
       
```

```python
# Specify the indices where you want to insert elements from Rating
index_to_insert = [7]

# Use a loop to insert elements from Rating into Ratings at the specified indices
for i, item in enumerate(Rating):
    Ratings.insert(index_to_insert[i], item)

# Print the updated Ratings
print(Ratings)
```

```python
len(Ratings)
```

```python
final_ratings=[]
for i in range(0,len(vehicles)):
    final_ratings.append(Ratings[i].get_text())
final_ratings    
```

```python
T_reviews=soup.find_all(class_="i-b fnt-12 lnk-c")
T_reviews
```

```python
total_reviews=[]

for i in range(0,len(vehicles)):
    total_reviews.append(T_reviews[i].get_text())

total_reviews    
```

```python
Total_Reviews=[]
for i in range(0,len(vehicles)):
    Total_Reviews.append(re.findall("[0-9]+[\s][\w]+",total_reviews[i]))
    
Total_Reviews    
```

```python
import pandas as pd
Tata_Cars=pd.DataFrame({'Car_Names':Car_Names,
                        'Engine|Mileage|fueltype':Car_info,
                        'Car_Prices':Car_Prices,
                        'Ratings':final_ratings ,
                        'Total_Reviews':Total_Reviews})
Tata_Cars
```

```python
Tata_Cars.to_csv("Tata_Cars.csv",index="false")
df=pd.read_csv("Tata_Cars.csv")
df
```
